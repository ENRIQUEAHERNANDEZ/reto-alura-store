### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()
```

```python
print("Ingreso total de Tienda 1:", ingreso_tienda1)
print("Ingreso total de Tienda 2:", ingreso_tienda2)
print("Ingreso total de Tienda 3:", ingreso_tienda3)
print("Ingreso total de Tienda 4:", ingreso_tienda4)
```

# 2. Ventas por categoría

```python
ventas_categoria_t1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria_t2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_t3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_t4 = tienda4["Categoría del Producto"].value_counts()
```

```python
print("Tienda 1 - Ventas por Categoría:")
print(ventas_categoria_t1)
print("\nTienda 2 - Ventas por Categoría:")
print(ventas_categoria_t2)
print("\nTienda 3 - Ventas por Categoría:")
print(ventas_categoria_t3)
print("\nTienda 4 - Ventas por Categoría:")
print(ventas_categoria_t4)
```

# 3. Calificación promedio de la tienda


```python
promedio_t1 = tienda["Calificación"].mean()
promedio_t2 = tienda2["Calificación"].mean()
promedio_t3 = tienda3["Calificación"].mean()
promedio_t4 = tienda4["Calificación"].mean()
```

```python
print("Promedio de calificación - Tienda 1:", promedio_t1)
print("Promedio de calificación - Tienda 2:", promedio_t2)
print("Promedio de calificación - Tienda 3:", promedio_t3)
print("Promedio de calificación - Tienda 4:", promedio_t4)
```

# 4. Productos más y menos vendidos

```python
ventas_producto_t1 = tienda["Producto"].value_counts()
ventas_producto_t2 = tienda2["Producto"].value_counts()
ventas_producto_t3 = tienda3["Producto"].value_counts()
ventas_producto_t4 = tienda4["Producto"].value_counts()
```

```python
# Tienda 1
print("Producto más vendido - Tienda 1:")
print(ventas_producto_t1.head(1))
print("Producto menos vendido - Tienda 1:")
print(ventas_producto_t1.tail(1))

# Tienda 2
print("\nProducto más vendido - Tienda 2:")
print(ventas_producto_t2.head(1))
print("Producto menos vendido - Tienda 2:")
print(ventas_producto_t2.tail(1))

# Tienda 3
print("\nProducto más vendido - Tienda 3:")
print(ventas_producto_t3.head(1))
print("Producto menos vendido - Tienda 3:")
print(ventas_producto_t3.tail(1))

# Tienda 4
print("\nProducto más vendido - Tienda 4:")
print(ventas_producto_t4.head(1))
print("Producto menos vendido - Tienda 4:")
print(ventas_producto_t4.tail(1))
```

# 5. Envío promedio por tienda

```python
envio_promedio_t1 = tienda["Costo de envío"].mean()
envio_promedio_t2 = tienda2["Costo de envío"].mean()
envio_promedio_t3 = tienda3["Costo de envío"].mean()
envio_promedio_t4 = tienda4["Costo de envío"].mean()
```

```python
print("Costo de envío promedio - Tienda 1:", envio_promedio_t1)
print("Costo de envío promedio - Tienda 2:", envio_promedio_t2)
print("Costo de envío promedio - Tienda 3:", envio_promedio_t3)
print("Costo de envío promedio - Tienda 4:", envio_promedio_t4)
```

##**Visulizacion de Graficos**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Ingresos por tienda
ingresos = pd.Series({
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
})

# Gráfico de barras
ingresos.plot(kind="bar", color="skyblue", title="Ingresos Totales por Tienda")
plt.ylabel("Ingreso Total ($)")
plt.xlabel("Tiendas")
plt.tight_layout()
plt.show()
```

```python
# Ventas por categoría en Tienda 1
categorias_t1 = tienda["Categoría del Producto"].value_counts()

# Gráfico de torta
categorias_t1.plot(kind="pie", autopct="%1.1f%%", startangle=90, figsize=(6,6))
plt.title("Distribución de Categorías - Tienda 1")
plt.ylabel("")  # Oculta la etiqueta de y
plt.tight_layout()
plt.show()
```

```python
# Asegurar que la columna de fecha esté en formato datetime
tienda["Fecha de Compra"] = pd.to_datetime(tienda["Fecha de Compra"])

# Agrupar por fecha y sumar ingresos (Precio)
ingresos_por_fecha = tienda.groupby("Fecha de Compra")["Precio"].sum()

# Graficar
plt.figure(figsize=(10,5))
ingresos_por_fecha.plot(kind="line", marker="o", color="teal")
plt.title("Ingresos Diarios - Tienda 1")
plt.xlabel("Fecha de Compra")
plt.ylabel("Ingreso Diario ($)")
plt.grid(True)
plt.tight_layout()
plt.show()

```

# ¡Extra! Análisis del desempeño geográfico

```python
plt.figure(figsize=(8,6))
plt.scatter(tienda["lon"], tienda["lat"], alpha=0.4, c='blue')
plt.title("Distribución geográfica de ventas - Tienda 1")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(
    x=tienda["lon"], y=tienda["lat"],
    cmap="Reds", fill=True, bw_adjust=0.5
)
plt.title("Mapa de calor de ventas - Tienda 1")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.tight_layout()
plt.show()
```

```python
# Crear un nuevo DataFrame con ubicación e ingreso
ubicacion_ventas = tienda.groupby(["lat", "lon"])["Precio"].sum().reset_index()

plt.figure(figsize=(8,6))
sns.scatterplot(
    x="lon", y="lat", size="Precio",
    hue="Precio", data=ubicacion_ventas,
    palette="viridis", legend=False, sizes=(20, 200)
)
plt.title("Ingresos por ubicación geográfica - Tienda 1")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.tight_layout()
plt.show()
```

```python
import folium

# Crear mapa centrado en una coordenada
m = folium.Map(location=[20, -100], zoom_start=5)

# Agregar marcadores por ubicación
for _, row in tienda.iterrows():
    folium.CircleMarker(
        location=[row["lat"], row["lon"]],
        radius=3,
        color="blue",
        fill=True,
        fill_opacity=0.6
    ).add_to(m)

m  # Mostrar mapa
```
